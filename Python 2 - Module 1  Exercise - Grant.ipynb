## Exercise 1: 
Create a Player class with a name, a position, and a speed attributes. Set the initial value of speed to 1 and create any necessary methods.
Then, 
1. Create two Player objects and print information about both players.
2. Assign 8 to player 1's speed
3. print information about both players.

````python
class Player() :
    """Player class has name, position, and speed attributes"""
    
    def __init__(self, name, position) :
        """Initialization method with name, position, and speed parameters"""
        self.name = name
        self.position = position
        self.speed = 1
    
    def print_info(self) :
        print(f"{self.name} plays {self.position} and has a speed of {self.speed}.")
        return
````

````python
player1 = Player("Bart", "defense")
player1.print_info()
player2 = Player("Lisa", "goalie")
player2.print_info()
print()
player1.speed = 8
player1.print_info()
player2.print_info()
````

## Exercise 2: 
Create a class called Rectangle with height and length attributes and has following methods: 
- perimeter() : returns a perimeter of a rectangle object
- area() : returns an area of a rectangle object

Then,
1. create two instances of Rectangle class
2. print each rectangle's perimeter and determine which rectangle has larger perimeter
3. print each rectangle's area and determine which rectangle has larger area

````python
class Rectangle() :
    """Rectangle class has height and length attributes"""
    
    def __init__(self, height, length) :
        """Initialization method with height and length parameters"""
        self.height = height
        self.length = length
    
    def area(self) :
        return (self.height * self.length)
    
    def perimeter(self) :
        return (2 * self.height + 2 * self.length)
````

````python
rect1 = Rectangle(6, 10)
rect2 = Rectangle(7, 9)
perimeter1 = rect1.perimeter()
perimeter2 = rect2.perimeter()
print(f"Rectangle 1 perimeter: {perimeter1}")
print(f"Rectangle 2 perimeter: {perimeter2}")
if perimeter1 > perimeter2 :
    print("Rectangle 1 has a greater perimeter")
if perimeter1 < perimeter2 :
    print("Rectangle 2 has a greater perimeter")
if perimeter1 == perimeter2 :
    print("Both rectangles have the same perimeter")
print()
area1 = rect1.area()
area2 = rect2.area()
print(f"Rectangle 1 area: {area1}")
print(f"Rectangle 2 area: {area2}")
if area1 > area2 :
    print("Rectangle 1 has a greater area")
if area1 < area2 :
    print("Rectangle 2 has a greater area")
if area1 == area2 :
    print("Both rectangles have the same area")
````

## Exercise 3: 
Create a class called Triangle with height and base attributes and has following methods:

- perimeter() : returns a perimeter of a right triangle object
- area() : returns an area of a right triangle object

Then,

- create two instances of Triangle class
- print each triangle's perimeter and determine which triangle has larger perimeter
- print each triangle's area and determine which triangle has larger area

Use 
- perimeter of a right triangle = height + base + Square-Root((height \** 2) + (base \** 2)) 
- area of a right triangle = ( height * base ) / 2

You'll need to <b>import math</b> library and use <b>math.sqrt()</b> function.



````python
import math

class Triangle() :
    """Rectangle class has height and base attributes"""
    
    def __init__(self, height, base) :
        """Initialization method with height and base parameters"""
        self.height = height
        self.base = base
    
    def area(self) :
        return (self.height * self.base / 2)
    
    def perimeter(self) :
        return (self.height + self.base + math.sqrt((self.height ** 2) + (self.base ** 2)))
````

````python
tri1 = Triangle(6, 10)
tri2 = Triangle(7, 9)
perimeter1 = tri1.perimeter()
perimeter2 = tri2.perimeter()
print(f"Triangle 1 perimeter: {perimeter1:.1f}")
print(f"Triangle 2 perimeter: {perimeter2:.1f}")
if perimeter1 > perimeter2 :
    print("Triangle 1 has a greater perimeter")
if perimeter1 < perimeter2 :
    print("Triangle 2 has a greater perimeter")
if perimeter1 == perimeter2 :
    print("Both triangles have the same perimeter")
print()
area1 = tri1.area()
area2 = tri2.area()
print(f"Triangle 1 area: {area1}")
print(f"Triangle 2 area: {area2}")
if area1 > area2 :
    print("Triangle 1 has a greater area")
if area1 < area2 :
    print("Triangle 2 has a greater area")
if area1 == area2 :
    print("Both triangles have the same area")
````

````python
class Movie() :
    """Employee class has name, year, and genre attributes"""
    
    def __init__(self, name, year, genre) :
        """Initialization method with name, year, and genre parameters"""
        self.name = name
        self.year = year
        self.genre = genre
    
    def get_genre(self) :
        return self.genre
    
    def get_name(self) :
        return self.name
    
    def get_year(self) :
        return self.year
````

````python
def add_movie(name, year, genre) :
    movie = Movie(name, year, genre)
    movies.append(movie)
    print(f"{movie.get_name()} was added.")
    return

def default_movies() :
    movie = Movie("Toy Story", 1995, "animated")
    movies.append(movie)
    movie = Movie("Lion King", 1994, "animated")
    movies.append(movie)
    movie = Movie("Frozen", 2013, "animated")
    movies.append(movie)
    return

def delete_movie(num) :
    del_movie = movies.pop(num - 1)
    print(f"{del_movie.get_name()} was deleted.")
    return

def display_menu() :
    print(f"""COMMAND MENU
    \tlist -- list all movies
    \tadd -- add a movie
    \tdel -- delete a movie
    \texit -- exit program\n""")
    return

def list_movies() :
    print("List of movies:")
    count = 1
    for movie in movies :
        name = movie.get_name()
        print(f"{count}. {name}")
        count += 1

movies = []

# add default movies to list
default_movies()

# introduce the database and display the menu
print("The Ultimate Movie Database\n")
display_menu()

command = input("What would you like to do? ")
while command != "exit" :
    if command == "list" :
        list_movies()
    elif command == "add" :
        movie_name = input("Which movie would you like to add? ")
        movie_year = int(input("What year did the movie come out? "))
        movie_genre = input("What is the genre of the movie? ")
        add_movie(movie_name, movie_year, movie_genre)
    elif command == "del" :
        del_num = int(input("Which movie would you like to delete? Enter the number: "))
        if del_num < len(movies) :
            delete_movie(del_num)
        else :
            continue
    command = input("\nWhat would you like to do? (list/add/del/exit) ")
print("Goodbye!")
````

````python
class Employee() :
    """Employee class has first name, last name, joined date, and active attributes"""
    
    def __init__(self, firstname, lastname, joined_date) :
        """Initialization method with first name, last name, joined date, and active parameters"""
        self.firstname = firstname
        self.lastname = lastname
        self.joined_date = joined_date
        self.active = True
    
    def get_active(self) :
        return self.active
    
    def get_firstname(self) :
        return self.firstname
    
    def get_joined_date(self) :
        return self.joined_date
    
    def get_lastname(self) :
        return self.lastname
````
